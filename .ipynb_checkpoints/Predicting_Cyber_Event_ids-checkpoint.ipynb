# Data Inspection & Tidying

First we'll take a look at the first 20 entries in the training data

```python
train_data.head(20)
```

Do get a better look at the data, we can take a look at different entries in each column, and see if we can find a way of sorting them into categories

```python
for col in train_data.columns:
    print("FEATURE: {}\nCLASSES: {}\n".format(col,train_data[col].unique()))
```

It looks like we can. to do this we'll strip usernames, domain names, host names, and destinations of their trailing digits.

```python
for df in [train_data,test_data,val_data]:
    df['UserType'] = df.UserName.str.split('(\d+)', expand=True)[0]
    df['UserType'].fillna('NONE',inplace = True)
    df['HostType'] = df.LogHost.str.split('(\d+)', expand=True)[0]
    df['HostType'].fillna('NONE',inplace = True)
    df['DomainType'] = df.DomainName.str.split('(\d+)', expand=True)[0]
    df['DomainType'].fillna('NONE',inplace = True)
    df['DestinationType'] = df.Destination.str.split('(\d+)', expand=True)[0]
    df['DestinationType'].fillna('NONE',inplace = True)
    df['ProcessType'] = df.ProcessName.str.split('(\d+)', expand=True)[0]
    df['ProcessType'].fillna('NONE',inplace = True)
    df['FailureReason'].fillna('NONE',inplace = True)
    df['LogonTypeDescription'].fillna('NONE',inplace = True)
    df['SourceType'] = df.Source.str.split('(\d+)', expand=True)[0]
    df['SourceType'].fillna('NONE',inplace = True)
```

We also know a trailing $ on a username indicates that the user is logging in to their local machine, so we can create a new column to see if they are a local user


```python
for df in [train_data,test_data,val_data]:
    df['islocal'] = df.UserName.apply(lambda un: 'yes' if str(un)[-1] == '$' else 'No')
```

fianlly we can see our data has an authentication column, and a status column. If the row is authenticated, with status 0x0, the authentication was sucessful, otherwise it failed authentication. we can thus create a new colum, isAuthenticated, with values 'None', 'Sucess' and 'Failed'


```python
def isauthenticated(row):
    if row.Status == '0x0':
        return 'Success'
    elif str(row.Status) == 'nan':
        return 'NONE'
    else:
        return 'Failed'

for df in [train_data,test_data,val_data]:
    
    df['isAuthenticated'] = df.apply(lambda row: isauthenticated(row),axis = 1)
```

This leaves us with our more workable data:

```python
train_data[['UserType', 'HostType', 'DestinationType', 'ProcessType','DomainType',
           'islocal','isAuthenticated','FailureReason','LogonTypeDescription','SourceType']].head(10)
```

# Training the Model

### Create our engineered dataframes 
We now need to engineer the data we want to pass into SciKitLearn's random forrest classifier as a series of one hot vectors. We'll create the to_one_hot function to do this, which will help filter out any feature classes in the test or validation sets which are not in the training set.

```python
cols = ['UserType', 'HostType', 'DestinationType', 'ProcessType','DomainType',
           'islocal','isAuthenticated','FailureReason','LogonTypeDescription','SourceType']
```

```python


def to_one_hot(df,columns):
    One_hot_types = [[feature for feature in train_data[col].unique() if feature != 'NONE' ] for col in columns]
    df_out = pd.DataFrame()
    for col,features in zip(columns,One_hot_types):
        for feature in features:
            df_out["{}_{}".format(col,feature)] = df[col].apply(lambda feat: 1 if str(feat) == feature else 0)
    return df_out
```

```python
train_eng,test_eng,val_eng= [to_one_hot(df,cols) for df in [train_data,test_data,val_data]]
```

We can now view our engineered dataframe to see the one hot vectors

```python
train_eng.head()
```

Finally we can fit and train our model

```python
classifier = RandomForestClassifier(random_state = 1, n_estimators = 20)
classifier.fit(train_eng,train_labels)
print("MODEL SCORE: {}".format(classifier.score(test_eng,test_labels)))
```

After Training our model we can see the model scores 91.3% on the test set. But can we do better? by playing around with the parameters, removing 'Domain Type' from the one hot vectors improves our score

```python
cols_new = ['UserType', 'HostType', 'DestinationType', 'ProcessType',
           'islocal','isAuthenticated','FailureReason','LogonTypeDescription','SourceType']
train_eng,test_eng,val_eng= [to_one_hot(df,cols_new) for df in [train_data,test_data,val_data]]


classifier = RandomForestClassifier(random_state = 1, n_estimators = 20)
classifier.fit(train_eng,train_labels)
print("MODEL SCORE: {}".format(classifier.score(test_eng,test_labels)))

```

We can see our score has now improved to 93.2%. Now that we are happy, we can validate our model against the validation dataset.

```python
print("MODEL VALIDATION SCORE: {}".format(classifier.score(val_eng,val_labels)))
```

### Performance For Individual Events

Sometimes you may only be interested in a particular event, and a models reliability for predicting different classes is NOT the same (necessarily). For example, if you predicted it wouldn't snow every day you'd actually be very accurate, but youd never manage to predict a snowy day (exceptionally poor recall). Suppose Event IDs 5 and 7 were of particular importance to us. For that we'd be intereseted in the performance of these metrics so we'll take a look at that here. 

```python
guesses = classifier.predict(test_eng)

print('-----------------------------------------------')
for number in [5,7]:
    Ground_Truth_bin = []
    Predict_bin = []
    for ground_truth, guess  in zip(list(test_labels),guesses):
        Ground_Truth_bin +=[1] if ground_truth == number else [0]
        Predict_bin +=[1] if guess == number else [0]

    print("Recall score for Event ID {} is : {}".format(number,recall_score(Ground_Truth_bin,Predict_bin)))
    print("Precision score for Event ID {} is : {}".format(number,precision_score(Ground_Truth_bin,Predict_bin)))
    print("Accuracy for Event ID {} is : {}".format(number,accuracy_score(Ground_Truth_bin,Predict_bin)))
    print("F1 score for Event ID {} is : {}".format(number,f1_score(Ground_Truth_bin,Predict_bin)))
    print('-----------------------------------------------')
```

From this analysis we can see Event ID 5 has very poor recall, which also means it has a low f1 score. This is because f1 score accounts for both recall and precision. If we take a closer look at the model, it falsely predicts Event ID 5s as 7s

```python
for ground_truth, guess  in zip(list(test_labels),guesses):
        if ground_truth == 5:
            print("GROUND TRUTH = 5 --> GUESS = {}".format(guess))
       
```
